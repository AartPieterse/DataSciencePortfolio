```python
import pandas as pd
pd.options.mode.chained_assignment = None

import seaborn as sns
```

```python
data = sns.load_dataset('penguins')
data.shape
```

## Correlations

```python
datacorr = data.corr()
datacorr.style.background_gradient(cmap='coolwarm', axis=None).format(precision=2)
```

There is strong positive correlation between Flipper_Length and Body_Mass. This is obvious because a big penguin needs big flippers to swim.

A negative correlation shows between Flipper_Length and Bill_Depth. So a penguin with a longer flipper has, according to these number, a smaller sized beak. 

Between Bill_Depth and Bill_Length is no correlation. It seems a penguins' beak is or long, or deep, and it doesn't grow equally bigger. 

## Scatterplots

### Strong positive correlation: flipper_length & body_mass

```python
data.plot(kind='scatter', x='flipper_length_mm', y='body_mass_g', figsize=(8, 5))
```

Bigger penguins have longer flippers, no surprises here.

### Strong negative correlation: flipper_length & bill_depth

```python
data.plot(kind='scatter', x='flipper_length_mm', y='bill_depth_mm', figsize=(8, 5))
```

There is a negative correlation, but this plot is more interesting.
There are two clouds visible and I think it's the division between two penguin species. One species with a big beak and short flippers, and another with a smaller beak and long flippers. 

### Weak correlation: bill_length & bill_depth

```python
data.plot(kind='scatter', x='bill_length_mm', y='bill_depth_mm', figsize=(8, 5))
```

The dots in this scatterplot look quite random, so we can conclude there is indeed not a correlation between bill length and bill depth.
